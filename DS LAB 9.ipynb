```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df
```

```python
sns.countplot(x='Age',data=df)
```

```python
sns.barplot(x='Age',y='Sex',data=df)
```

```python
sns.boxplot(x='Age',y='Sex',hue='Survived',data=df)
```

```python
sns.pairplot(data=df)
```

```python
sns.jointplot(x='Age',y='Pclass',hue='Survived',data=df)
```

```python
sns.histplot(x='Age',y='Sex',data=df)
```

```python
sns.countplot(df)
```

```python
sns.scatterplot(df)
```

```python
sns.catplot(df)
```

```python
sns.boxenplot(df)
```
